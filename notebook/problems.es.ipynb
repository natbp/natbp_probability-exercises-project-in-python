## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random

def simulacion(it):
    
    count = 0
    
    for i in range(it):
        dado1 = random.randint(1, 6)
        dado2 = random.randint(1, 6)
        suma = dado1 + dado2
        
        if (suma > 7) or (suma % 2 == 0):
            count += 1
    return count / it

print("Probabilidad estimada:", simulacion(1000))
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            

def simulacion_bolas(it):
    
    case1 = 0
    case2 = 0
    for i in range(it):
        blanca = 0
        roja = 0
        verde = 0
        
        for j in range(5):
            bola = random.choice(list(ball_box.values()))
            
            if bola == "White":
                blanca += 1
            elif bola == "Red":
                roja += 1
            else:
                verde += 1
        
        if blanca == 3 and roja == 2:
            case1 += 1
        
        if blanca == 5 or roja == 5 or verde == 5:
            case2 += 1
            
    probCase1 = case1 / it
    probCase2 = case2 / it
        
    return probCase1, probCase2
    
diferentes, iguales = simulacion_bolas(1000)

print("La probabilidad de que salgan 3 bolas blancas y 2 rojas es de", diferentes)
print("La probabilidad de que salgan las 5 bolas del mismo color es de",iguales)
```

## Ejercicio 1
*Qué es combinatoria, permutación y como esto se relaciona con probabilidad.*

- Combinatoria: Cuántas formas diferentes existen de organizar, seleccionar o agrupar elementos

        comb = n! / r!(n-r)!

- Permutación: Nos determina el número total de formas en que se pueden ordenar u organizar los elementos. El orden aquí SÍ importa. 

        perm = n! / (n-r)!

*¿Cómo se relacionan?*

    La combinatoria y la permutación permiten contar el número total de resultados posibles y favorables, lo cual es fundamental para calcular probabilidades teniendo en cuenta que la formula es casos_favorables / casos_posibles.

***Ejercicio 1.1***

*Diferencia entre en combinatoria o permutación.*

    La diferencia entre la permutación y la combinatoria es que durante un intercambio la posición de los elementos nos va a producir un resultado totalmente diferente ya que para la permutación el orden sí existe


```python
from scipy.special import comb, factorial
import math
```

```python
def combinacion(n, r):
    return math.factorial(n) // (math.factorial(r) * math.factorial(n - r))
```

```python
def permutacion(n, r):
    return math.factorial(n) // math.factorial(n - r)
```

*Un entrenador dispone de 22 jugadores para formar un equipo de fútbol.
¿Cuántas alineaciones de 11 jugadores puede hacer?*


```python
#Opción 1
alineaciones = comb(22, 11, exact= True)
print(alineaciones)

#Opción 2
alineaciones = math.comb(22, 11)
print(alineaciones)

#Opción 3
print("Alineaciones:", combinacion(22, 11))
```
